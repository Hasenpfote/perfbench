```python
import numpy as np
from perfbench.process import *
```

```python
bm = Benchmark(
    datasets=[
        Dataset(
            stmt=lambda n: np.random.uniform(low=-1., high=1., size=n).astype(np.float64), 
            title='float64'
        )
    ],
    dataset_sizes=[2 ** n for n in range(15)],
    kernels=[
        Kernel(
            stmt=lambda x: np.around(x), 
            label='around'
        ),
        Kernel(
            stmt=lambda x: np.rint(x), 
            label='rint'
        )
    ],
    xlabel='dataset sizes',
    title='around vs rint'
)
bm.run()
```

```python
bm.save_as_png(filepath='./temp.png')
```
